```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation, NMF
```

```python
texts = hf['content'].fillna('')
```

```python
hf
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
tfidf = tfidf_vectorizer.fit_transform(texts)
```

```python
nmf = NMF(n_components=5, random_state=42)
nmf_topics = nmf.fit_transform(tfidf)
```

```python
def print_topics(model, feature_names, n_top_words=10):
    for topic_idx, topic in enumerate(model.components_):
        top_features = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        print(f"Topic {topic_idx + 1}: {' | '.join(top_features)}")
```

```python
print("🔹 NMF Topics:")
print_topics(nmf, tfidf_vectorizer.get_feature_names_out())
```

```python
def plot_distribution(probs):
    st.subheader("🔍 Topic Probability Distribution")
    fig, ax = plt.subplots()
    ax.bar(range(1, len(probs)+1), probs)
    ax.set_xlabel("Topic")
    ax.set_ylabel("Probability")
    ax.set_xticks(range(1, len(probs)+1))
    st.pyplot(fig)

def get_top_topic(probs):
    return int(probs.argmax()) + 1
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Load your data
df = pd.read_csv("sports_data.csv")  # Adjust the path if needed

# Fill missing content
texts = df['content'].fillna('')

# Vectorize using CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_features=1000)
X = vectorizer.fit_transform(texts)

# Fit LDA model
lda = LatentDirichletAllocation(n_components=5, random_state=42)
topic_distributions = lda.fit_transform(X)

# Assign most probable topic to each article
df['Most_Probable_Topic'] = topic_distributions.argmax(axis=1) + 1  # Topic numbers start from 1

# View results
print(df[['headlines', 'Most_Probable_Topic']].head())

# (Optional) Save to CSV
df.to_csv("final_project_with_topics.csv", index=False)
```
